```python
# Importing libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome("C:/Program Files/Google/Chrome/Application/chromedriver.exe")
driver.get("https://chia-anime.su/anime/list-mode/")

anime_names = driver.find_elements(By.CSS_SELECTOR, ".series")
animes = []
for names in anime_names:
    anime_url = names.get_attribute("href")
    animes.append(anime_url)
```

```python
# Creating empty lists
anime = []
rating_list = []
others = []

# Creating a for loop that will loop through a series of items in the website
for i in range (len(animes)):
    driver.get(animes[i])
    driver.refresh()
    rating_elems = driver.find_element(By. CSS_SELECTOR, ".rating")
    l = driver.find_element(By.CSS_SELECTOR, ".entry-title")
    s = driver.find_element(By.CSS_SELECTOR, 'div.spe')
    
    anime.append(l.text)
    rating_list.append(rating_elems.text)
    others.append(s.text)
        
anime

driver.close()
```

The other list has a lot of information. The information was scraped from the site in this manner because there wasn't a way to scrape the individual elements (i.e rating, year, and studio). In the next stage of this process, this list will be split into three separate lists for rating, release year and studio.

```python
str_list = []
for f in others:
    str(f)
    str_list.append(str(f)) 
str_list
```

```python
r_list = []
for r in str_list:
    r_list.append(r.split("\n"))
r_list
```

```python
# Spliting the studios
studio = []
for sl in r_list:
    studio.append(sl[1].split(":")[1])
studio
```

```python
# Spliting the years
released = []
for rl in r_list:
    released.append(rl[2].split(":")[1])
released
```

If you observe the year list, you will notice that some entries have the wrong format. This will be taken care of during the data cleaning process.


```python
# Spliting the ratings
rating = []
for rt in rating_list:
    rating.append(str(rt).split()[1])
rating
```

```python
# Creating a dictionary to combine the lists
import pandas as pd

data = {'name': anime,
        'studio': studio,
        'yearReleased': released,
        'rating': rating}

# Changing dictionary to pandas dataframe
new = pd.DataFrame.from_dict(data)
  
new
```

```python
# Saving dataframe as csv
new.to_csv('AnimeList.csv')
```
